```python
with open('../input/D04.txt', 'r') as f:
    data = [[[x for x in range(int(i.split('-')[0]), int(i.split('-')[1])+1)] for i in e.split(',')] for e in f.read().splitlines()]
data
```

```python
sum((set(i[1]).issubset(set(i[0])) | set(i[0]).issubset(set(i[1]))) for i in data)
```

```python
sum(True if set(i[1]).intersection(set(i[0])) else False for i in data)
```
