```python
tracker = {}
X = 1
cycle = 0
tracker[cycle] = X
for cmd in data:
    if cmd[0] == 'noop':
        cycle += 1
    if cmd[0] == 'addx':
        cycle += 1
        tracker[cycle] = X
        cycle += 1
        X += int(cmd[1])
    tracker[cycle] = X
tracker
```
