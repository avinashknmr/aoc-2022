```python
import re
import numpy as np
from copy import deepcopy
```

```python
with open('../input/D05.txt', 'r') as f:
    data = f.read().splitlines()
raw_stack = [[x[i] for x in data[:8] if x[i].isalpha()][::-1] for i in np.arange(1,37,4)]
steps = [[int(x) for x in re.findall('\d+', s)] for s in data[10:]]
raw_stack, steps
```

```python
def get_final_stack_order(stack, steps, engine=9000):
    global raw_stack
    for n, f, t in steps:
        c = stack[f-1][-1*n:]
        c = c if engine==9001 else c[::-1]
        del stack[f-1][-1*n:]
        stack[t-1] += c
    return ''.join([s[-1] for s in stack])
```

```python
get_final_stack_order(deepcopy(raw_stack), steps, 9000)
```
