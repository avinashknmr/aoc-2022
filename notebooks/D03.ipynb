```python
with open('../input/D03.txt', 'r') as f:
    data = f.read().splitlines()
data
```

```python
import string
trans = {a: i+1 for i,a in enumerate(list(string.ascii_letters))}
trans
```

```python
error_items = [list(set(b[:len(b)//2]).intersection(set(b[len(b)//2:])))[0] for b in data]
sum(trans[i] for i in error_items)
```

```python
import numpy as np
elf_groups = np.array(data).reshape((len(data)//3, 3))
```

```python
badges = [list(set(b[0]) & set(b[1]) & set(b[2]))[0] for b in elf_groups]
sum(trans[i] for i in badges)
```
